# Tensorflow y keras para el entrenamiento de Redes Neuronales 

## Librerias necesarias

```python
import tensorflow as tf
import tensorflow_datasets as tfds
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
```

```python
import h5py
```

```python
def loadData(nameH5py):
    trainDataSet = h5py.File(f'./{nameH5py}.h5', 'r')
        
    trainSetX = np.array(trainDataSet["trainX"][:])
    trainSetY = np.array(trainDataSet["trainY"][:])
    testSetX = np.array(trainDataSet["testX"][:])
    testSetY = np.array(trainDataSet["testY"][:])
    classes = np.array(trainDataSet["classes"][:], dtype='|S7')

    trainSetY = trainSetY.reshape((1, trainSetY.shape[0]))
    testSetY = testSetY.reshape((1, testSetY.shape[0]))

    return trainSetX, trainSetY, testSetX, testSetY, classes
```

```python
trainSetX, trainSetY, testSetX, testSetY, classes = loadData("data")
trainSetX.shape, trainSetY.shape
```

```python
trainSetX = trainSetX.reshape(trainSetX.shape[0], 64 * 64 * 3)
trainSetY = np.squeeze(trainSetY)
```

```python
trainSetY
```

```python
num_rows = 10
num_cols = 10
num_images = num_rows * num_cols

plt.figure(figsize=(num_cols, num_rows))
for i in range(num_images):
    plt.subplot(num_rows, num_cols, i + 1)
    plt.imshow(trainSetX[i])
    plt.xticks([])
    plt.yticks([])

plt.tight_layout()
plt.show()
```

```python
modelGenre = tf.keras.models.Sequential([
  tf.keras.layers.Input(shape=(64, 64, 3)),  # Input layer especificando la forma de los datos de entrada
  tf.keras.layers.Flatten(),                 # Capa Flatten
  tf.keras.layers.Dense(256, activation='relu'), 
  tf.keras.layers.Dense(128, activation='relu'), 
  tf.keras.layers.Dense(64, activation='relu'), 
  tf.keras.layers.Dense(10, activation='sigmoid')
])
modelGenre.compile(optimizer='sgd', # Decenso del gradiente
  loss='sparse_categorical_crossentropy', # Funcion de perdida
  metrics=['accuracy']) 

modelGenre.fit(trainSetX, trainSetY, epochs=20) #Numero de iteraciones en este caso 20
modelGenre.evaluate(testSetX, testSetY) #resultado tanto del porcentaje de acertividad como el de la perdida
```

## Dataset

```python
mnist = tf.keras.datasets.mnist
```

```python
(x_train, y_train),(x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
```

```python
x_train.shape, y_train.shape
```

```python
y_train
```

## Ver el data set visualmente

```python
num_rows = 10
num_cols = 10
num_images = num_rows * num_cols

plt.figure(figsize=(num_cols, num_rows))
for i in range(num_images):
    plt.subplot(num_rows, num_cols, i + 1)
    plt.imshow(x_train[i], cmap='gray')
    plt.xticks([])
    plt.yticks([])

plt.tight_layout()
plt.show()
```

## Entrenamiento

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(28, 28)), #poner todos los datos de entrada de dimension (28, 28, )
  tf.keras.layers.Dense(256, activation='relu'), #Primera capa oculta donde la funcion de activacion es Relu con 256 neuronas
  tf.keras.layers.Dense(128, activation='relu'), #Segunda capa oculta donde la funcion de activacion es Relu con 128 neuronas
  tf.keras.layers.Dense(128, activation='relu'), #Tercera capa oculta donde la funcion de activacion es Relu con 128 neuronas
  tf.keras.layers.Dense(10, activation='softmax') #Capa de salida donde la funcion de activacion es Softmax con una salida de 10 posibles resultados
])

model.compile(optimizer='sgd', # Decenso del gradiente
  loss='sparse_categorical_crossentropy', # Funcion de perdida
  metrics=['accuracy']) 

model.fit(x_train, y_train, epochs=20) #Numero de iteraciones en este caso 20
model.evaluate(x_test, y_test) #resultado tanto del porcentaje de acertividad como el de la perdida
```

## Porcentaje de acertividad

```python
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print(test_acc)
```

```python
predictions = model.predict(x_test)
predicted_labels = np.argmax(predictions, axis=1)

print(f"Clase predicha: {predicted_labels[0]}")
print(predicted_labels)
```

## Datos para visualizar de manera grafica los resultados

```python
confusion = confusion_matrix(y_test, predicted_labels)

sns.heatmap(confusion, annot=False, cmap='Blues')
plt.xlabel('Clase Predicha')
plt.ylabel('Clase Real')
plt.show()
```

```python

plt.figure(figsize=(12, 12))
for i in range(25):
    plt.subplot(5, 5, i + 1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(x_test[i], cmap=plt.cm.binary)
    if predicted_labels[i] == y_test[i]:
        color = 'blue'
    else:
        color = 'red'
    plt.xlabel(f"Real: {y_test[i]}\nPredicho: {predicted_labels[i]}", color=color)
plt.show()
```
